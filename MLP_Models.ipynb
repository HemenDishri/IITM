```python
lookup_common_dict = \
{
                "isn't"	:	"is not"	,
                "aren't"	:	"are not"	,
                "wasn't"	:	"was not"	,
                "weren't"	:	"were not"	,
                "haven't"	:	"havent"	,
                "hasn't"	:	"has not"	,
                "hadn't"	:	"had not"	,
                "won't"	:	"wont"	,
                "wouldn't"	:	"would not"	,
                "don't"	:	"dont"	,
                "doesn't"	:	"doesnt"	,
                "didn't"	:	"didnt"	,
                "can't"	:	"cant"	,
                "couldn't"	:	"could not"	,
                "shouldn't"	:	"should not"	,
                "mightn't"	:	"might not"	,
                "mustn't"	:	"must not"	,
                "would've"	:	"would have"	,
                "should've"	:	"should have"	,
                "could've"	:	"could have"	,
                "might've"	:	"might have"	,
                "must've"	:	"must have"
}
```

```python
def lookup_words(input_text):
    words = input_text.split()
    new_words = []
    for word in words:
        if word.lower() in lookup_common_dict:
            word = lookup_common_dict[word.lower()]
        new_words.append(word)
    new_text = " ".join(new_words)
    return new_text
```

```python
def noise_removal(text):
    #Standardizing words first
    std_text = lookup_words(text)


    #Getting list of stopwords from nltk
    basic_stop_words = stopwords.words('english')
    to_take = ['not','very', 'nor', 'too','more','most']  # can be useful in differentiating between 1,2,3,4,5
    for i in to_take:
        basic_stop_words.remove(i)

    #Adding some extra stopwords which people generally use which writing a review
    add_stop_words = ['hi', 'hii','hiii', 'lol', 'omg', 'oh', 'ohh', 'hello', 'yo', 'etc', 'btw', "i", "he", "him", "it"]
    stop_words = basic_stop_words + add_stop_words + ["the", "is", "are" ,"a", "by", "am", "for", "in","to", "as", "of", "at", "an","be"]


    #1. Lower casing the words
    std_text = std_text.lower()

    #2. Removing white spaces
    space_free_text = std_text.strip()

    #3. Removing non-ascii character
    ascii_free_text = re.sub('[^\x00-\x7f]',' ', space_free_text)

    #4. Removing emojis, punctuations and numbers
    pure_word_text = re.sub('[^a-zA-Z]', ' ', ascii_free_text)

    emoji_pattern = re.compile("["
                       u"\U0001F600-\U0001F64F"  # emoticons
                       u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                       u"\U0001F680-\U0001F6FF"  # transport & map symbols
                       u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                       u"\U00002702-\U000027B0"
                       u"\U000024C2-\U0001F251"
                       "]+", flags=re.UNICODE)

    emoji_free_txt = emoji_pattern.sub(r'', pure_word_text)

    #5. Splitting text into words
    text_words = emoji_free_txt.split() # text is spliited into list of words

    #6. Removing stopwords and stemming the word to root level
    ps = PorterStemmer()
    stopword_free_word = [ps.stem(word) for word in text_words if not word in set(stop_words)] # Removing stopwards by applying loop.

    #7. joining the words together to get clean_text
    clean_review = ' '.join(stopword_free_word)
    return clean_review
```

```python
# prompt: write belmaan ford algorithm

def bellman_ford(graph, source):
  """
  Bellman-Ford algorithm for finding shortest paths from a single source.

  Args:
    graph: A dictionary representing the graph, where keys are nodes and values
           are lists of tuples (neighbor, weight).
    source: The source node.

  Returns:
    A dictionary containing the shortest distances from the source to all other nodes.
  """

  # Initialize distances to infinity
  distances = {node: float('inf') for node in graph}
  distances[source] = 0

  # Relax edges repeatedly
  for _ in range(len(graph) - 1):
    for node in graph:
      for neighbor, weight in graph[node]:
        if distances[node] + weight < distances[neighbor]:
          distances[neighbor] = distances[node] + weight

  # Check for negative cycles
  for node in graph:
    for neighbor, weight in graph[node]:
      if distances[node] + weight < distances[neighbor]:
        raise ValueError("Graph contains a negative cycle")

  return distances

# Example usage:
graph = {
    'A': [('B', -1), ('C', 4)],
    'B': [('C', 3), ('D', 2), ('E', 2)],
    'C': [],
    'D': [('B', 1), ('C', 5)],
    'E': [('D', -3)]
}

source_node = 'A'
shortest_distances = bellman_ford(graph, source_node)

print("Shortest distances from node", source_node, ":")
for node, distance in shortest_distances.items():
  print(node, ":", distance)
```

```python
# prompt: how nested for loops works

# Nested for loops are simply for loops within for loops. The outer loop executes first, and for each iteration of the outer loop, the inner loop executes completely.

# Example: Printing a multiplication table
for i in range(1, 6):  # Outer loop (rows)
  for j in range(1, 6):  # Inner loop (columns)
    print(i * j, end="\t")  # Print the product, separated by tabs
  print()  # Move to the next line after each row

# Explanation:

# 1. Outer Loop: The outer loop iterates from 1 to 5 (inclusive), representing the rows of the multiplication table.

# 2. Inner Loop: For each value of 'i' (each row), the inner loop iterates from 1 to 5 (inclusive), representing the columns.

# 3. Calculation and Output: Inside the inner loop, the product of 'i' and 'j' is calculated and printed, followed by a tab to separate the values in the same row.

# 4. Newline: After the inner loop completes for a given 'i' (row), a newline character is printed to move to the next row.

# This process continues until both loops have finished iterating, resulting in a 5x5 multiplication table being printed to the console.

# Key Points:

# - The inner loop executes completely for each iteration of the outer loop.
# - The variables of the outer loop can be accessed within the inner loop.
# - Nested loops can be used to iterate over multi-dimensional data structures like matrices or tables.
```

```python
# prompt: count the maximum number of decreasing sequence in a list and ignore
#  if not in seequence and list can be increasing or decreasing

def count_decreasing_sequences(lst):
  count = 0
  decreasing = False
  for i in range(1, len(lst)):
    if lst[i] < lst[i - 1]:
      if not decreasing:
        count += 1
        decreasing = True
    else:
      decreasing = False
  return count

# Example usage
lst1 = [5, 4, 3, 2, 1]
lst2 = [1, 2, 3, 2, 1]
lst3 = [1, 2, 3, 4, 5]

print(count_decreasing_sequences(lst1))  # Output: 1
print(count_decreasing_sequences(lst2))  # Output: 1
print(count_decreasing_sequences(lst3))  # Output: 0
```
